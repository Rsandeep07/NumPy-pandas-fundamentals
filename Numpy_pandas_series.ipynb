1. Create a NumPy array with numbers from 1 to 20.
   * a) Print the array.
   * b) Reshape it into a 4x5 matrix.
   * c) Print only the even numbers from the array.

```python
import numpy as np

# 1. Create a NumPy array with numbers from 1 to 20
arr = np.arange(1, 21)

# a) Print the array
print("Original array:",arr)


# b) Reshape it into a 4x5 matrix
reshaped_arr = arr.reshape(4, 5)
print("\nReshaped to 4x5 matrix:")
print(reshaped_arr)

# c) Print only the even numbers from the array
even_numbers = arr[arr % 2 == 0]
print("\nEven numbers in the array:")
print(even_numbers)
```

5. Create a NumPy array with 100 values ranging from 0 to 1 (both inclusive) using linspace.
   * a) Find the sum of all values less than 0.5.
   * b) Count how many values are greater than 0.8.

```python
import numpy as np

# Create a NumPy array with 100 values ranging from 0 to 1 (both inclusive) using linspace
arr = np.linspace(0, 1, 100)
print("Array created using linspace:")
print(arr)
print("----"*20)
# a) Find the sum of all values less than 0.5.
sum_val = np.sum(arr[arr < 0.5])
print("sum of all values less than 0.5. is",sum_val)
# Count how many values are greater than 0.8.
val_greater = np.sum(arr[arr>0.8]) 
print("Values greater than 0.8 is",val_greater)
```

   * c) Add a new subject "Computer" with score 89.

#### 8. Create two Series of integers with some overlapping values.
   * a) Find the union, intersection, and difference of both Series.
   * b) Compare both Series element-wise for equality.

```python
import pandas as pd

# Create two Series of integers with some overlapping values
series_1 = pd.Series([1, 2, 3, 4, 5])
series_2 = pd.Series([4, 5, 6, 7, 8])

print("Series 1:")
print(series_1)

print("\nSeries 2:")
print(series_2)

# Find the union, intersection, and difference
union = pd.Series(list(set(series_1).union(set(series_2))))
intersection = pd.Series(list(set(series_1).intersection(set(series_2))))
difference = pd.Series(list(set(series_1).difference(set(series_2))))

print("\nUnion of series1 and series2:")
print(union)

print("\nIntersection of series1 and series2:")
print(intersection)

print("\nDifference (series1 - series2):")
print(difference)

# Element-wise equality comparison
# (Only valid if they have the same length)
if len(series_1) == len(series_2):
    elementwise_comparison = series_1 == series_2
else:
    # Align indexes and fill missing values
    s1_aligned, s2_aligned = series_1.align(series_2, fill_value=None)
    elementwise_comparison = s1_aligned == s2_aligned

print("\nElement-wise equality comparison:")
print(elementwise_comparison)
```

#### 9. Create a Series with custom indexes (e.g., days of the week).
   * a) Access the values for "Monday" and "Wednesday".
   * b) Slice from "Tuesday" to "Friday".
   * c) Change the value for "Sunday" to 100.

```python
#  Create a Series with custom indexes (e.g., days of the week).
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
values = [10, 20, 30, 40, 50, 60, 70]
week_series = pd.Series(values, index=days)
print(week_series)
print("---"*25)
#Access the values for "Monday" and "Wednesday".
print("Access the values for Monday and Wednesday")
print(week_series[['Monday', 'Wednesday']])
# Slice from "Tuesday" to "Friday".
print("Slice from Tuesday to Friday:")
print(week_series['Tuesday':'Friday'])
# c) Change the value for "Sunday" to 100.
week_series['Sunday'] = 100
print("Series after changing Sunday to 100:")
print(week_series)
```

#### 10. Create a Series of 10 random numbers using np.random.randn().
    * a) Calculate the mean and standard deviation.
    * b) Replace all negative values with 0.
    * c) Plot the Series using .plot() (optional if plotting covered).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a Series of 10 random numbers from standard normal distribution
series = pd.Series(np.random.randn(10))

print("Original Series:")
print(series)

# a) Calculate the mean and standard deviation
mean_val = series.mean()
std_val = series.std()

print("\nMean:", mean_val)
print("Standard Deviation:", std_val)

# b) Replace all negative values with 0
series[series < 0] = 0
print("\nSeries after replacing negative values with 0:")
print(series)

# c) Plot the Series (optional)
series.plot(kind='line', marker='o', title="Random Series Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

#### 11. What is the difference between a NumPy array and a Pandas Series?

* NumPy array is a fixed-size, homogeneous, multi-dimensional array for numerical computations.
* Pandas Series is a one-dimensional labeled array that can hold data with custom indexes.
* NumPy arrays use integer-based positional indexing, while Pandas Series use label-based indexing.
* Pandas Series can handle missing data (NaN) more gracefully than NumPy arrays.
* Pandas Series is built on top of NumPy arrays, adding metadata and data analysis features.

#### 13. How is broadcasting handled differently in NumPy vs Pandas?

**NumPy Broadcasting:**
* Works purely on array shapes without labels.
* Matches dimensions from the end and stretches dimensions of size 1 to fit.
* If shapes are incompatible after broadcasting rules, raises a ValueError.

**Pandas Broadcasting:**
* Considers both shapes and index/column labels.
* Aligns data based on labels first, then broadcasts values.
* If labels dont match, missing positions are filled with NaN before broadcasting.

```python
When you add two Pandas Series with different indexes, Pandas aligns them by their index labels (not positions).

If a label is present in both Series, their values are added.

If a label is missing in one Series, the result is NaN for that label.
```
